```python
import math
import matplotlib.pyplot as plt

list_pos = [(51.5028, -0.2801), (51.4951, -0.2547), (51.4936, -0.2251), (51.4905, -0.2139),
            (51.492, -0.1973), (51.4945, -0.1829), (51.4941, -0.1738), (51.4924, -0.1565),
            (51.4965, -0.1447), (51.5067, -0.1428), (51.515, -0.1415), (51.5165, -0.131),
            (51.5174, -0.12), (51.523, -0.1244)]

def cal_dis(pos1, pos2):
    return math.sqrt((pos1[0] - pos2[0])**2 + (pos1[1] - pos2[1])**2)

list_dis = [cal_dis(list_pos[i], list_pos[i + 1]) for i in range(len(list_pos) - 1)]

mean_distance = sum(list_dis) / len(list_dis)

plt.bar(range(len(list_dis)), list_dis)
plt.axhline(y=mean_distance, color='r', linestyle='-', label='Mean Distance')
plt.xlabel('Edge Index')
plt.ylabel('Distance')
plt.legend()
plt.show()
```
