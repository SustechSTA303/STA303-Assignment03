```python
import matplotlib.pyplot as plt

# 数据
algorithms = ['setA','setB','setC','setD']
setA = [0.000283, 0.000501, 0.001097, 0.002889]
setB = [0.000135, 0.000823, 0.001265, 0.003555]
setC = [0.000110, 0.000558, 0.001687, 0.005072]
setD = [0.000089, 0.000503, 0.001706, 0.004896]
weightedA = [0.000146, 0.000462, 0.000934, 0.001645]

# 绘制折线图
plt.plot(algorithms, setA, label='Dijkstra')
plt.plot(algorithms, setB, label='A* with Euclidean Distance')
plt.plot(algorithms, setC, label='A* with Manhattan Distance')
plt.plot(algorithms, setD, label='A* with Chebyshev distance')
plt.plot(algorithms, weightedA, label='Weighted A*')

# 添加标签和标题
plt.xlabel('set')
plt.ylabel('Time (seconds)')
plt.title('Algorithm Comparison')
plt.legend()

plt.savefig('picture/algorithm_comparison.png')
# 显示图形
plt.show()



```
