```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('./time.csv')
data
```

```python
Astar = data.iloc[0:10, 1:]

fig, ax = plt.subplots(figsize=(6, 4))


ax.plot(Astar.iloc[:,0], 'o--', label='route 1')
ax.plot(Astar.iloc[:,1], 'o--', label='route 2')
ax.plot(Astar.iloc[:,2], 'o--', label='route 3')

ax.set_ylabel('seconds')
ax.set_title('Astar')

ax.legend(fontsize='small')
plt.xticks(visible=False)
plt.tight_layout()

plt.savefig('./imagedata/Astar.png', dpi=300, bbox_inches='tight')

plt.show()

a1 = np.mean(Astar.iloc[:,0])
a2 = np.mean(Astar.iloc[:,1])
a3 = np.mean(Astar.iloc[:,2])

print(a1,a2,a3)
```

```python
Dijkstra = data.iloc[11:21, 1:]

fig, ax = plt.subplots(figsize=(6, 4))


ax.plot(Dijkstra.iloc[:,0], 'o--', label='route 1')
ax.plot(Dijkstra.iloc[:,1], 'o--', label='route 2')
ax.plot(Dijkstra.iloc[:,2], 'o--', label='route 3')

ax.set_ylabel('seconds')
ax.set_title('Dijkstra')

ax.legend(fontsize='small')
plt.xticks(visible=False)
plt.tight_layout()

plt.savefig('./imagedata/Dijkstra.png', dpi=300, bbox_inches='tight')

plt.show()

d1 = np.mean(Dijkstra.iloc[:,0])
d2 = np.mean(Dijkstra.iloc[:,1])
d3 = np.mean(Dijkstra.iloc[:,2])

print(d1,d2,d3)
```

```python
SPFA = data.iloc[22:32, 1:]

fig, ax = plt.subplots(figsize=(6, 4))


ax.plot(SPFA.iloc[:,0], 'o--', label='route 1')
ax.plot(SPFA.iloc[:,1], 'o--', label='route 2')
ax.plot(SPFA.iloc[:,2], 'o--', label='route 3')

ax.set_ylabel('seconds')
ax.set_title('SPFA')
plt.xticks(visible=False)
ax.legend(fontsize='small')
plt.xticks(visible=False)
plt.tight_layout()

plt.savefig('./imagedata/SPFA.png', dpi=300, bbox_inches='tight')

plt.show()

s1 = np.mean(SPFA.iloc[:,0])
s2 = np.mean(SPFA.iloc[:,1])
s3 = np.mean(SPFA.iloc[:,2])

print(s1,s2,s3)

```

```python
BF = data.iloc[33:, 1:]

fig, ax = plt.subplots(figsize=(6, 4))


ax.plot(BF.iloc[:,0], 'o--', label='route 1')
ax.plot(BF.iloc[:,1], 'o--', label='route 2')
ax.plot(BF.iloc[:,2], 'o--', label='route 3')

ax.set_ylabel('seconds')
ax.set_title('Bellman-Ford')

ax.legend(fontsize='small')

plt.tight_layout()
plt.xticks(visible=False)
plt.savefig('./imagedata/BF.png', dpi=300, bbox_inches='tight')

plt.show()

b1 = np.mean(BF.iloc[:,0])
b2 = np.mean(BF.iloc[:,1])
b3 = np.mean(BF.iloc[:,2])

print(s1,s2,s3)
```

```python
A=[a1,a2,a2]
D=[d1,d2,d3]
S=[s1,s2,s3]
B=[b1,b2,b3]
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(A, 'o--', label='Astar')
ax.plot(D, 'o--', label='Dijkstra')
ax.plot(S, 'o--', label='SPFA')
ax.plot(B, 'o--', label='Bellman-Ford')
ax.set_ylabel('seconds')
ax.legend(fontsize='small')
plt.tight_layout()
labels=['route 1','route 2','route 3']
plt.xticks(range(len(labels)), labels)
plt.savefig('./imagedata/total.png', dpi=300, bbox_inches='tight')
```
