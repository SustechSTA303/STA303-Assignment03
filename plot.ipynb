```python
import seaborn as sns
import matplotlib.pyplot as plt
b10=[]
b20=[0.721]
b30=[2.851]


b1=[3.815]
b2=[0.896]
b3=[3.706]


b12=[]
b22=[0.865]
b32=[2.920]



sns.set(style="whitegrid")  # 设置样式，可选
plt.figure(figsize=(4, 6))  # 设置图形大小，可选

# 将列表合并成一个二维列表
data0= [b10,b20,b30]
data = [b1,b2,b3]
data2= [b12,b22,b32]
methods=['UCS', 'Greedy Best-First', 'A*']
patitions=['L1','L2','Che']

# 绘制图形
fig, ax = plt.subplots()
colors = ['blue', 'green', 'red'] 
marker_size=7

for i, row in enumerate(data0):
    ax.plot([i-0.1] * len(row), row, 'o',color=colors[0],markersize=marker_size)
ax.add_artist(plt.Line2D([0], [0], color=colors[0], marker='o', label=patitions[0]))  # 添加第一组点的图例项

for j, row in enumerate(data):
    ax.plot([j] * len(row), row, 'o',color=colors[1],markersize=marker_size)
ax.add_artist(plt.Line2D([0], [0], color=colors[1], marker='o', label=patitions[1]))  # 添加第一组点的图例项

for k, row in enumerate(data2):
    ax.plot([k+0.1] * len(row), row, 'o',color=colors[2],markersize=marker_size)
ax.add_artist(plt.Line2D([0], [0], color=colors[2], marker='o', label=patitions[2]))  # 添加第一组点的图例项

# 设置横轴和纵轴标签
ax.set_xlabel('search method')
ax.set_ylabel('hehuristic cost')

# 设置横轴刻度和标签
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(methods)

# 添加图例并设置图例字体大小
legend = ax.legend()
for label in legend.get_texts():
    label.set_fontsize('7')  # 设置图例文字的字体大小

# 显示图形
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 创建数据框
data = {
    'Time': ["UCS",
    "Greedy Best-First",
    "A*"],
    'L1': [
    None,0.721,1.233
],
    'L2':[
    3.815,0.896,2.435
] ,
    'Che': [
    None,0.865,1.657
]
}

df = pd.DataFrame(data)

# 设置样式
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

# 绘制柱状图
ax = sns.barplot(data=df.melt(id_vars='Time', var_name='Method', value_name='Value'), x='Time', y='Value', hue='Method')

# 添加标题和标签
plt.title('Time Consumptions')
plt.xlabel('Search algorithm')
plt.ylabel('Time consumption(s)')

# 设置图例位置
plt.legend(loc='upper left')

# 旋转 x 轴标签
plt.xticks(rotation=10)

# 显示图形
plt.show()
```
