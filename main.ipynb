```python
import matplotlib.pyplot as plt

categories = ['A* (Euclidean)', 'A* (Manhattan)', 'A* (Zero)']
values = [0.0001003742218017578, 6.008148193359375e-05, 5.7697296142578125e-05]

# 画柱状图
plt.bar(categories, values)

# 添加标题和标签
plt.title('variant heuristic functions for A* algorithm')
plt.xlabel('Categories')
plt.ylabel('Values')

# 显示图表
plt.show()
```
