```python
import matplotlib.pyplot as plt

# 示例数据
execution_times = {
    'A* (Euclidean)': 0.00013065338134765625,
    'Bellman-Ford': 0.03273200988769531,
    'Dijkstra': 0.0316927433013916
}

# 绘制执行时间的柱状图
plt.figure(figsize=(8, 6))
bars = plt.bar(execution_times.keys(), execution_times.values(), color='skyblue')

# 添加数据标签
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval * 0.5, round(yval, 5), ha='center', va='bottom', color='black', fontsize=9)

plt.xlabel('Algorithm')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time of Different Algorithms\nStart: Acton Town,End: Turnham Green')
plt.ylim(0, max(execution_times.values()) * 1.1)  # 设置纵轴范围略大于最大值，以显示文本信息
plt.show()
```
