```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

algorithms = ['A*', 'UCS', 'Greedy', 'Bidirectional A*']
path_lengths_manhattan = [0.30839,0.30839,0.34473,0.30994]
path_lengths_euclidean = [19.855629,19.855629,22.165082,19.994634]
path_lengths_degree = [0.250079,0.250079,0.270522,0.251589]
path_lengths_nodes = [14.9,14.9,23.5,15.1]
path_lengths_manhattan = [element * 100 for element in path_lengths_manhattan]
path_lengths_degree = [element * 100 for element in path_lengths_degree]

sns.set(style="whitegrid")

fig, ax = plt.subplots(figsize=(10, 6))

barWidth = 0.2
r1 = np.arange(len(algorithms))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

# Using Seaborn to plot the differences in path lengths with transposed axes and a different color palette
colors = sns.color_palette('tab20b', 4)

ax.barh(r1, path_lengths_manhattan, color=colors[0], height=barWidth, label='100*Manhattan')
ax.barh(r2, path_lengths_euclidean, color=colors[1], height=barWidth, label='Euclidean')
ax.barh(r3, path_lengths_degree, color=colors[2], height=barWidth, label='100*Degree')
ax.barh(r4, path_lengths_nodes, color=colors[3], height=barWidth, label='Nodes')

ax.set_ylabel('Algorithms')
ax.set_xlabel('Difference in Path Lengths')
ax.set_title('Difference in Path Lengths by Algorithm and Distance Calculation')
ax.set_yticks([r + barWidth * 1.5 for r in range(len(algorithms))])
ax.set_yticklabels(algorithms)
ax.legend()

plt.tight_layout()
plt.show()
```
